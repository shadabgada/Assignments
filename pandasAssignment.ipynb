```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'], 
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],  
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
df
```

```python
#Question no 1
missing = np.where(df['FlightNumber'].isnull())
for i in missing[0]:
    df['FlightNumber'][i]=df['FlightNumber'][i-1]+10
df['FlightNumber']=df.FlightNumber.astype(int)
df
```

```python
#Question no 2
temp=pd.DataFrame(df.From_To.str.split('_').tolist(),columns=['From','To'])
temp
```

```python
#Question no 3
temp['From']=temp['From'].str.title()
temp['To']=temp['To'].str.title()
temp
```

```python
#Question no 4
df=pd.concat([temp,df],axis=1)
del df['From_To']
df
```

```python
#Question no 5
n=df['RecentDelays'].str.len().max()
col_names=[]
for i in range(n):
    col_names.append('delay_{}'.format(i))
col_names
```

```python
#Question no 5

m=df['RecentDelays'].size
a = np.empty((m,n))
a[:] = np.nan
delay=pd.DataFrame(data=a,columns=col_names)
for i,val1 in enumerate(df['RecentDelays']):
    for j,val2 in enumerate(val1):
        delay[col_names[j]][i]=val2        

df=pd.concat([df,delay],axis=1)
del df['RecentDelays']
df
```
